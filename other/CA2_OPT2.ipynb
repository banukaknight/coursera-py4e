```python
print ('''Q1. Write a Python program to take 5 lists as an input and sum
all the elements of list, if it does not contain any even elements. 
In case a particular list contains any even elements then elements of that
list should not be considered in the overall sum of all lists.\n ''')

#this function calculate sum of a given list
#if list had any even elements: returns 0. if no even elements present: return sum.
def getSum_SkipEven(myList):
    thisListSum = 0 #variable to hold sum of this List
    listHasEven = False #flag to signal if list has even values
    getSum_SkipEven.counter +=1 #static variable-counter used to check which list is being processed
    
    for item in myList:
        if item%2 == 0:
            #As soon as loop come across an even value listHasEven-flag will be set to True
            listHasEven = True
        thisListSum += item
    
    if(listHasEven):
        print("List",getSum_SkipEven.counter ,"had even values. Sum Ignored:",thisListSum)
        return 0
    else:
        print("List",getSum_SkipEven.counter ,"ONLY had odd values. Sum Returned:",thisListSum)
        return thisListSum
    # end of getSum_SkipEven() function
    
#Declare 5 lists hardcoded data
list1 = [3,115,963,1995,5,7887] #contain only odd
list2 = [88,64,12,676,88,972,222] #contain even
list3 = [111,65,8,3452,78,23,67] #contain odd and even mix
list4 = [333,767,811,915,55,779,235] #contain only odd
list5 = [125,71,779,869,241,119,587] #contain only odd
    
getSum_SkipEven.counter = 0 #static variable Used to check which list is being processed
allListSum = 0 #variable to store sum of all lists having no even values

print("-- 5 Lists with hardcoded data:")
for thisList in (list1,list2,list3,list4,list5):
    print (thisList)

print("-- Calculate sum of each list:")
for thisList in (list1,list2,list3,list4,list5):
    thisListSum = getSum_SkipEven(thisList)
    allListSum += thisListSum
    
print ("\nSum of lists that had no even numbers is:",allListSum)

print("-------------------------------\n-------------------------------")

listOfLists = [] #list to hold 5 lists of dynamic user data

print("Get Data from user and build 5 dynamic lists containing integers")
for i in range (0,5):
    list_length = int(input("Enter length for list " + str(i+1) +": ") )
    tempList = []
    for j in range(0, list_length):
        intInput = int(input("Data for list " + str(i+1) +": ") )
        tempList.append(intInput)
        
    listOfLists.append(tempList)
    
getSum_SkipEven.counter = 0 #static variable Used to check which list is being processed
allListSum = 0 #variable to store sum of all lists having no even values

print("\n--5 Lists with Data from user:")
for thisList in listOfLists:
    print (thisList)

print("-- Calculate sum of each list:")

for thisList in listOfLists:
    thisListSum = getSum_SkipEven(thisList)
    allListSum += thisListSum
    
print ("\nSum of lists that had no even numbers is:",allListSum)
```

```python
print ('''Q2. Write a Python program to find second largest element 
and second smallest element in a tuple.\n''')

import math #math module to use positive/negative infinity

#function getLargest() return largest value in given tuple
def getLargest(myTuple):
    largest = -math.inf #using negative infinity as default largest
    for i in myTuple:
        if i > largest:
            largest = i
    return largest

#function getSecondLargest() return second largest value in given tuple
def getSecondLargest(myTuple):
    largest = getLargest(myTuple) #call getLargest() function
    secondLargest = -math.inf
    for i in myTuple:
        if i > secondLargest and i < largest:
            secondLargest = i
    return secondLargest

#function getSmallest() return smallest value in given tuple
def getSmallest(myTuple):
    smallest = math.inf #using positive infinity as default smallest
    for i in myTuple:
        if i < smallest:
            smallest = i
    return smallest

#function getSecondSmallest() return second smallest value in given tuple
def getSecondSmallest(myTuple):
    smallest = getSmallest(myTuple) #call getSmallest() function
    secondSmallest = math.inf
    for i in myTuple:
        if i < secondSmallest and i > smallest:
            secondSmallest = i
    return secondSmallest

tuple2 = (111,543,-24,65,235,-17,676,42,54,-53,12,6)
print("Tuple concidered: ",tuple2)

print("\nSecond largest element in tuple is:", getSecondLargest(tuple2) )
print("Second smallest element in tuple is:", getSecondSmallest(tuple2) )

print("-------------------------------\n")
print("Get Data from user and build a tuple with dynamic data:")
tuple_length = int(input("Enter length for desired tuple:"))
tuple1 = tuple()
for i in range (0,tuple_length):
    tempValue = int(input("Enter Values for tuple:") )
    tuple1 = tuple1 + (tempValue,)

print("\nTuple with dynamic data: ",tuple1)

print("Second largest element in tuple is:", getSecondLargest(tuple1) )
print("Second smallest element in tuple is:", getSecondSmallest(tuple1) )
```

```python
print('''Q3. Write a Python program to get the maximum and minimum value in a dictionary.''')
import math

def getMaxAndMin(dict1):
    maxkeys = []
    maxval = -math.inf #set maxval to negative infinity
    minkeys = []
    minval = math.inf #set minval to positive infinity

    for key in dict1:
        if dict1[key] == maxval: #new key with current maxval found
            maxkeys.append(key) #append key to maxkeys-list
        elif dict1[key] > maxval: #new max value found
            maxval = dict1[key] #re-assign new maxval
            maxkeys = [key] #rest maxkeys-list.

        if dict1[key] == minval: #new key with current minval found
            minkeys.append(key) #append key to minkeys-list
        elif dict1[key] < minval: #new min value found
            minval = dict1[key] #re-assign new minval
            minkeys = [key] #rest minkeys-list.
    print("\nDictionary concidered:\n", dict1)        
    print("\n--Maximum value found:",maxval)
    print("Max value subject/s:",maxkeys)
    print("--Minimum value found:",minval)
    print("Min value subject/s:",minkeys)
    # end of getMaxAndMin() function
    
#declare a dictionary with hardcoded data
student1_marks = {
    "Maths":75,
    "Biology":45,
    "Chemistry":89,
    "Physics":89,
    "Social Science":50,
    "Sports":89
}

#call getMaxAndMin() function to process the dictionary with hardcoded data    
getMaxAndMin(student1_marks) 
    
#create dictionary dynamically with user input
subject_count = int(input("\nEnter how many subjects taken by student:"))
student2_marks = {} #emty dictionary to hold user input

for i in range(0,subject_count):
    sub_name = input("Name of subject "+ str(i+1)+ ": ")
    sub_marks = int(input("Enter marks for- "+sub_name+ " between 0 & 100: "))
    if sub_marks >= 0 and sub_marks <=100:
        #insert data to dictionary only-if marks are between 0-100
        student2_marks[sub_name] = sub_marks
        
#call getMaxAndMin() function to process the dictionary with dynamic data    
getMaxAndMin(student2_marks) 
```
