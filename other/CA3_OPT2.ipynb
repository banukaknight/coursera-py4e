```python
print ('''Q1. Write a Python program to take input of 10 numbers, find 
the square of all 10 numbers and check whether the square of every number 
is palindrome or not. Numbers where square is palindrome, save in file 
with original numbers and their squares. Numbers where square is 
not palindrome, must be saved in separate file.\n''')

import datetime  #datetime module used to get current system time.
import os  #os module used to check for file exisitence.

TEN = 10  #Control number of inputs accepted in program
FILE1 = "11917550_only_palindromes.txt" #preset file names
FILE2 = "11917550_not_palindromes.txt" #so it can be modified from one place

# isPalindrome() - CHECKS FOR PALINDROME
# (using python-ternary-op, int-string convertion & str-reverese)
def isPalindrome(numb):
    return True if (numb == int( str(numb)[::-1] )) else False

#getSqure() - Return squre of a number
def getSqure(numb):
    return numb**2

#lambda / annonymous function. (can place this instead of getSqure f-call)
#getSqure2 = lambda numb: numb**2  

# #uncomment for Testing purpose only
# input_list = [7,9,11,5,2,-5,-10,100,4,-111]

input_list = []  #this list will hold the input from user
while (len(input_list)<TEN):
    temp_msg = 'Enter Number ' + str(len(input_list) +1) + ' : '
    try:
        temp_int = int(input(temp_msg))
        input_list.append(temp_int)
    except Exception as e:
        print("***Error found: ",e)
        print("**Expecting Integer value. Please try again.")

#calculate square of each number form input_list[] and map to a new square_list[]
squre_list = list( map(getSqure, input_list) )
        
print("List of Inputs: ", input_list)
print("List of square: ", squre_list)

fh1 = open(FILE1, "w")  #open file in 'w' mode where any old content is cleared from file
fh2 = open(FILE2, "w")
#write a starting line in the files.
fh1.write(FILE1 + ": created at: " + str(datetime.datetime.now()) + "\n___Numbers for which Square is Palindrome.\n")
fh2.write(FILE2 + ": created at: " + str(datetime.datetime.now()) + "\n___Numbers for which Square is not Palindrome.\n")

#traverese list by index and write to file based on isPalindrome() function result
for xx in range(len(input_list)):
    temp_str = "{} squared is: {}\n".format(input_list[xx], squre_list[xx])
    if isPalindrome(squre_list[xx]):
        fh1.write(temp_str)
    else:
        fh2.write(temp_str)

fh1.close()  #close filehead that used to write to file
fh2.close()

print("\nDisplay Two text File contents: \n")
for fname in [FILE1,FILE2]:
    if os.path.exists(fname):
        f = open(fname, "r")
        print(f.read())
        f.close()
    
print("Finished execution at: ",datetime.datetime.now())
```

```python
print ('''Q2. Write a Python program to calculate % of death rate, 
happened due to Covid-19 in all over INDIA with the help of function. 
% of death rate must be returned by the function. All required values 
must be entered on run time like total cases, no. of deaths etc.\n''')

import datetime  #datetime module used to get current system time.

## getDeathRate() - returns "Infection Fatality Rate (IFR)" = Deaths / Cases 
def getDeathRate(totD,totC):
    try:
        #expect ZeroDivisionError incase total-Cases or population is set to 0 by user
        return (totD/totC)*100
    except:
        return 0

## getInfectedPr() - returns "Presentage of population infected with covid-19"
def getInfectedPr(totC,pop):
    try:
        return (totC/pop)*100
    except:
        return 0

## getRecoveryRate() - returns "Recovery rate of patient once infected"
def getRecoveryRate(totR,totC):
    try:
        return (totR/totC)*100
    except:
        return 0

country = 'India'
resource = "https://www.worldometers.info/coronavirus/"
dataRetrievedOn = datetime.datetime.now()
##default static data - will be used if user input has any error.
population = 1384678624
totalCases = 8364086
totalDeaths = 124354
totalRecovered = 7711809
resetData = False

print("Please enter integer values as input. \nIf user input is Erroneous, program will use default static data for calclulations.\n")
try:
    ##try to get dynamic data input from user
    country = input("Enter Country: ")
    population = int(input("Enter Population of the Country: "))
    totalCases = int(input("Enter Total covid-19 positive cases recorded sofar: "))
    totalDeaths = int(input("Enter Total covid-19 deaths recorded in india: "))
    totalRecovered = int(input("Enter Total covid-19 recovered patients: "))
except Exception as e:
    print("***Error found: ",e)
    print("\nProgram will use default data stored to calculate results.")
    resetData = True
finally:
    ##cleanup code makes sure user input is acceptable, if not reset data.
    if(population<0 or totalCases<0 or totalRecovered<0):
        print("Expecting positive integer values for input.")
        resetData = True
    elif(population<totalCases):
        print("Total covid cases recorded must be less than actual population!")
        resetData = True
    elif( totalCases<(totalDeaths+totalRecovered) ):
        print("Sum of total deaths and recovered cannot be larger than Total covid confirmed cases!")
        resetData = True
        
    if(resetData):
        print("\nUser has entered bad data. Program will use default data stored to calculate results.")
        country = 'India'
        population = 1380004385
        totalCases = 8335273
        totalDeaths = 123783
        totalRecovered = 7675269
    
##call user defined functions to calculate deathRate and other values.
deathRate = getDeathRate(totalDeaths, totalCases)
infectedPresentage = getInfectedPr(totalCases, population)
recoveryRate = getRecoveryRate(totalRecovered, totalCases)
activeCasesRate = 100-deathRate-recoveryRate

print("\nAccording to Data retrieved from: {} \nOn the date: {}".format(resource,dataRetrievedOn) )
print("{}, a country of {:,} population had total of {:,} confirmed Covid-19 cases.".format(country,population,totalCases) )
print("There has been {:,} covid related deaths, and {:,} patients recovered.".format(totalDeaths,totalRecovered) )

print("\nIn {}, Covid-19 Death Rate (Infection Fatality Rate) is recorded as: {:.4f}%".format(country, deathRate))
print("\n{:.4f}% of {}'s population has contracted the virus with a recovery rate of {:.4f}%".format(infectedPresentage,country,recoveryRate))


print('\n***Attempt to create pie chart to better display data.***')
import matplotlib.pyplot as plotter # import the pyplot library

print("Result of Covid cases recoreded in:", country)
pieLabels= 'Covid Infection Fatality Rate%', 'Covid Recovery Rate%', 'Active Cases Rate%'
dataforpie = [deathRate, recoveryRate, activeCasesRate ]
figureObject, axesObject = plotter.subplots()
# Draw the pie chart
axesObject.pie(dataforpie, labels=pieLabels, autopct='%1.2f', startangle=90)

axesObject.axis('equal') # Aspect ratio - equal means pie is a circle
plotter.show()
```
